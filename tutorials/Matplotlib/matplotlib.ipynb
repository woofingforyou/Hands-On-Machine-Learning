```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import Image
import numpy as np

# Dane - wierzchołki kwadratu
x = [-2, 0, 0, -2, -2]
y = [2, 2, -2, -2, 2]

fig, ax = plt.subplots()
ax.set_xlim(-3, 1)
ax.set_ylim(-3, 3)
ax.set_title("Animated Square with Point Labels")

line, = ax.plot([], [], marker='o', color='b')
text_labels = []

# Inicjalizacja wykresu
def init():
    line.set_data([], [])
    return line,

# Krok animacji
def update(i):
    line.set_data(x[:i+1], y[:i+1])
    for txt in text_labels:
        txt.remove()
    text_labels.clear()
    for j in range(i+1):
        label = ax.text(x[j], y[j] + 0.3, f'({x[j]},{y[j]})', ha='center')
        text_labels.append(label)
    return line, *text_labels

# Tworzenie animacji
anim = animation.FuncAnimation(fig, update, frames=len(x), init_func=init,
                               blit=True, interval=600, repeat=False)

# Zapis jako GIF
anim.save("square.gif", writer='pillow', fps=1)

# Wyświetlenie w notebooku
Image(filename="square.gif")
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 300)
fig, (ax1, ax2, ax3) = plt.subplots(3, 1)

ax1.plot(x, np.sin(x), "r")
ax2.plot(x, np.cos(x), "b")
ax3.plot(x, np.tan(x), "g")
ax3.set_ylim(-5, 5)

ax1.set_title("sin")
ax2.set_title("cos")
ax3.set_title("tan")

for ax in (ax1, ax2, ax3):
    ax.grid(True)

fig.set_size_inches(10, 8)
fig.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 10, 300)

fig, axs = plt.subplots(2,2, sharex=True)

ax1 = axs[0, 0]
ax2 = axs[0, 1]
ax3 = axs[1, 0]
ax4 = axs[1, 1]

line1,line2 = ax1.plot(np.sin(x),"r", np.cos(x), "b")
ax1.legend([line1,line2],['sin = red','cos = blue'])

line3, = ax2.plot(np.exp(-x/3) * np.sin(5*x))
ax2.grid(True)
line3.set_color("violet")

line4, = ax3.plot(np.sqrt(x), "g")
ax3.set_xlabel("x")
ax3.set_ylabel("square root of x")

line5, = ax4.plot(np.log(x + 1))
line5.set_color("orange")
ax4.set_ylim(0,3)
ax4.set_title("log(x+1)")

fig.set_size_inches(10,8)
fig.tight_layout()
plt.show()
```
